```python
import pandas as pd
import glob
import matplotlib.pyplot as plt

kgs_to_bpm = 1 / 2.65

# Step 1: Use glob to find all files matching the pattern
file_pattern = 'lhs_run_out_sub*.csv'
files = glob.glob(file_pattern)

# Step 2: Initialize a list to store dataframes
dataframes = []

# Step 3: Loop through the files and read them into pandas DataFrames
for file in files:
    df = pd.read_csv(file)
    frac_aperture = df["param_frac_aperature"].iloc[0]
    frac_roughness=  df["param_frac_roughness"].iloc[0]
    plt.plot(df["time"],df["p_in_1"], label=f"{frac_aperture:.1E},{frac_roughness:.1E}")
plt.plot(df["time"], df["relative_pressure_16a"],"k--", label="experiment")
plt.legend()
plt.show()
    # dataframes.append(df)

# Step 3: Loop through the files and read them into pandas DataFrames
for file in files:
    df = pd.read_csv(file)
    frac_aperture = df["param_frac_aperature"].iloc[0]
    matrix_poro=  df["param_matrix_poro"].iloc[0]
    plt.plot(df["time"],df["fluid_report"] / df["a1_dt"] * kgs_to_bpm, label=f"{frac_aperture:.1E},{matrix_poro:.1E}")
plt.plot(df["time"], df["mass_rate"],"k--", label="experiment")
plt.legend()
```
